```python
# Imports/Setup
import os, gc, math, json, time, random, warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, roc_auc_score, roc_curve, precision_recall_fscore_support, confusion_matrix
)
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

print('TensorFlow:', tf.__version__)

DATA_DIR = './dataset'
TRAIN_TRANSACTION = os.path.join(DATA_DIR, 'train_transaction.csv')
TRAIN_IDENTITY   = os.path.join(DATA_DIR, 'train_identity.csv')

assert os.path.exists(TRAIN_TRANSACTION), f'Arquivo no encontrado: {TRAIN_TRANSACTION}'
assert os.path.exists(TRAIN_IDENTITY),   f'Arquivo no encontrado: {TRAIN_IDENTITY}'

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

# Carregar & Mesclar
print("Carregando CSVs...")
train_transaction = pd.read_csv(TRAIN_TRANSACTION, low_memory=False)
train_identity    = pd.read_csv(TRAIN_IDENTITY, low_memory=False)

print('Shapes:', train_transaction.shape, train_identity.shape)

df = train_transaction.merge(train_identity, on='TransactionID', how='left')
print('Merged shape:', df.shape)

print(df.head())
print(df[['isFraud','TransactionDT','TransactionAmt']].describe(include='all'))

#EDA  distribuio, faltantes, correlao
# Distribuio do alvo
target_counts = df['isFraud'].value_counts(dropna=False).sort_index()
print('\nContagem alvo:\n', target_counts)
print('Taxa de fraude:', df['isFraud'].mean())

plt.figure()
target_counts.plot(kind='bar')
plt.title('Distribuio de isFraud')
plt.xlabel('isFraud')
plt.ylabel('contagem')
plt.show()

# Faltantes (top 30)
missing = df.isnull().mean().sort_values(ascending=False)
print("\nTop-30 colunas com mais faltantes:\n", missing.head(30))

plt.figure()
missing.head(30).plot(kind='bar')
plt.title('Top-30 faltantes')
plt.xlabel('coluna')
plt.ylabel('taxa')
plt.tight_layout()
plt.show()

# Distribuio TransactionAmt
if 'TransactionAmt' in df.columns:
    plt.figure()
    df['TransactionAmt'].dropna().hist(bins=100)
    plt.title('Distribuio de TransactionAmt')
    plt.xlabel('TransactionAmt')
    plt.ylabel('freq')
    plt.show()

    plt.figure()
    np.log1p(df['TransactionAmt'].dropna()).hist(bins=100)
    plt.title('Distribuio log1p(TransactionAmt)')
    plt.xlabel('log1p(TransactionAmt)')
    plt.ylabel('freq')
    plt.show()

# Proxy temporal: TransactionDT
if 'TransactionDT' in df.columns:
    plt.figure()
    df['TransactionDT'].dropna().hist(bins=100)
    plt.title('Distribuio de TransactionDT (segundos)')
    plt.xlabel('TransactionDT')
    plt.ylabel('freq')
    plt.show()

# Cardinalidade categrica
categorical_cols = [c for c in df.columns if df[c].dtype=='object']
card = pd.Series({c: df[c].nunique(dropna=True) for c in categorical_cols}).sort_values(ascending=False)
print("\nCardinalidade (top 20):\n", card.head(20))

# Correlao  subset numrico para no estourar memria
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
sample_numeric = numeric_cols[:40] if len(numeric_cols) > 40 else numeric_cols
corr = df[sample_numeric].corr(numeric_only=True)

plt.figure(figsize=(10, 8))
plt.imshow(corr, aspect='auto')
plt.colorbar()
plt.title('Heatmap de correlao (subset)')
plt.xticks(range(len(sample_numeric)), sample_numeric, rotation=90, fontsize=6)
plt.yticks(range(len(sample_numeric)), sample_numeric, fontsize=6)
plt.tight_layout()
plt.show()

# Outliers via IQR em TransactionAmt
if 'TransactionAmt' in df.columns:
    amt = df['TransactionAmt'].dropna()
    if len(amt) > 0:
        Q1, Q3 = np.percentile(amt, [25, 75])
        IQR = Q3 - Q1
        lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
        outlier_rate = ((amt < lower) | (amt > upper)).mean()
        print(f'\nLimites IQR TransactionAmt: {lower:.2f}, {upper:.2f}')
        print(f'Taxa de outliers (TransactionAmt): {outlier_rate:.4f}')

# Preparao de dados para LSTM (sequncias)

for col in ['card1','addr1','dist1']:
    if col not in df.columns:
        df[col] = np.nan

uid = (
    df['card1'].astype('Int64').astype(str) + '_' +
    df['addr1'].astype('Int64').astype(str) + '_' +
    df['dist1'].astype('Int64').astype(str)
)
df['uid'] = uid.values

# Subconjunto enxuto de colunas (ajuste se quiser incluir mais V*, C*, D*, etc.)
basic_num = ['TransactionDT', 'TransactionAmt', 'dist1', 'C1', 'C2', 'C5', 'D1', 'D10', 'V12', 'V13']
basic_num = [c for c in basic_num if c in df.columns]

basic_cat = ['ProductCD', 'card4', 'card6', 'P_emaildomain', 'R_emaildomain']
basic_cat = [c for c in basic_cat if c in df.columns]

target_col = 'isFraud'
use_cols = list(dict.fromkeys([*basic_num, *basic_cat, 'uid', 'TransactionDT', target_col]))
use_cols = [c for c in use_cols if c in df.columns]
work = df[use_cols].copy()

print('\nColunas de trabalho:', work.columns.tolist())
print('Shape work:', work.shape)

# Tratamento de faltantes
for c in basic_num:
    if c in work.columns:
        med = work[c].median()
        work[c] = work[c].fillna(med)

cat_maps = {}
for c in basic_cat:
    if c in work.columns:
        work[c] = work[c].fillna('missing').astype(str)
        cats = work[c].unique().tolist()
        cat_maps[c] = {v:i for i,v in enumerate(cats)}
        work[c] = work[c].map(cat_maps[c]).astype(np.int32)

# Normalizao de numricas (incluindo TransactionDT aqui, opcional)
scaler = StandardScaler()
scale_cols = [c for c in basic_num if c in work.columns]
if len(scale_cols) > 0:
    work[scale_cols] = scaler.fit_transform(work[scale_cols])

# Ordenar por uid/tempo
work = work.sort_values(['uid','TransactionDT']).reset_index(drop=True)

# Construo de sequncias
SEQ_LEN = 10   # tamanho da janela
STEP    = 1    # passo
feature_cols = [c for c in work.columns if c not in ['uid','TransactionDT', target_col]]

X_seq, y_seq = [], []

def add_sequences(group):
    values = group[feature_cols].values
    labels = group[target_col].values
    n = len(group)
    # janela inclusiva: i..i+SEQ_LEN-1
    for i in range(0, n-SEQ_LEN+1, STEP):
        X_seq.append(values[i:i+SEQ_LEN])
        y_seq.append(labels[i+SEQ_LEN-1])  # rtulo = ltimo elemento da janela

unique_uids = work['uid'].unique()
np.random.shuffle(unique_uids)
max_users = 5000  # ajuste conforme RAM/tempo; aumentar => mais dados
sel_uids = unique_uids[:max_users]

for u in sel_uids:
    g = work[work['uid']==u]
    if len(g) >= SEQ_LEN:
        add_sequences(g)

X_seq = np.array(X_seq, dtype=np.float32)
y_seq = np.array(y_seq, dtype=np.int32)

print('X_seq shape:', X_seq.shape, ' y_seq shape:', y_seq.shape)
print('Fraude (seq):', y_seq.mean())

# Split train / test
X_train, X_test, y_train, y_test = train_test_split(
    X_seq, y_seq, test_size=0.2, random_state=RANDOM_STATE, stratify=y_seq
)
print('Train:', X_train.shape, 'Test:', X_test.shape)
print('Fraude (train):', y_train.mean(), ' Fraude (test):', y_test.mean())

#Modelo LSTM
def build_lstm_model(input_shape, lstm_units=64, dense_units=64, dropout_rate=0.3):
    model = models.Sequential()
    model.add(layers.Input(shape=input_shape))
    model.add(layers.Masking())  # til se zeros representarem "vazio"
    model.add(layers.LSTM(lstm_units, return_sequences=True))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.LSTM(lstm_units//2))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(dense_units, activation='relu'))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='binary_crossentropy',
        metrics=[tf.keras.metrics.AUC(curve='ROC', name='AUC')]
    )
    return model

input_shape = (X_train.shape[1], X_train.shape[2])
model = build_lstm_model(input_shape, lstm_units=64, dense_units=64, dropout_rate=0.3)
model.summary()

# Treinamento Pesos de classe para desbalanceamento
pos = y_train.sum()
neg = len(y_train) - pos
# evita diviso por zero
weight_for_0 = (1.0 / max(neg, 1)) * (len(y_train)) / 2.0
weight_for_1 = (1.0 / max(pos, 1)) * (len(y_train)) / 2.0
class_weights = {0: weight_for_0, 1: weight_for_1}
print('Class weights:', class_weights)

es = callbacks.EarlyStopping(monitor='val_AUC', mode='max', patience=3, restore_best_weights=True)
rlr = callbacks.ReduceLROnPlateau(monitor='val_AUC', mode='max', factor=0.5, patience=2, min_lr=1e-5)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=15,
    batch_size=256,
    class_weight=class_weights,
    callbacks=[es, rlr],
    verbose=1
)

# Curvas de aprendizado
h = history.history

plt.figure()
plt.plot(h['loss'], label='train_loss')
plt.plot(h['val_loss'], label='val_loss')
plt.title('Curva de Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.figure()
plt.plot(h['AUC'], label='train_AUC')
plt.plot(h['val_AUC'], label='val_AUC')
plt.title('Curva de AUC')
plt.xlabel('Epoch')
plt.ylabel('AUC')
plt.legend()
plt.show()

# Avaliao no Teste
y_prob = model.predict(X_test).ravel()
y_pred = (y_prob >= 0.5).astype(int)

auc = roc_auc_score(y_test, y_prob)
prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary', zero_division=0)
print(f"\nAUC-ROC: {auc:.4f}")
print(f"Preciso: {prec:.4f}  Recall: {rec:.4f}  F1: {f1:.4f}")
print('\nClassification Report:\n', classification_report(y_test, y_pred, digits=4, zero_division=0))

# Curva ROC
fpr, tpr, thr = roc_curve(y_test, y_prob)
plt.figure()
plt.plot(fpr, tpr)
plt.title('ROC Curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()

# Matriz de confuso
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', cm)
plt.figure()
plt.imshow(cm)
plt.title('Confusion Matrix')
plt.xlabel('Predito')
plt.ylabel('Verdadeiro')
for (i,j), val in np.ndenumerate(cm):
    plt.text(j, i, int(val), ha='center', va='center')
plt.show()

```

# Relatrio  IEEE-CIS Fraud Detection com LSTM

## 1. Preparao dos Dados

- **Dataset**: `train_transaction.csv` (590.540  394) e `train_identity.csv` (144.233  41).
- Merge realizado via `TransactionID`, resultando em **590.540 linhas  434 colunas**.
- **Problemas identificados**:
  - Forte **desbalanceamento** do alvo: apenas ~3,5% das transaes so fraude.
  - Muitas variveis `id_*`, `D*`, `V*` com **mais de 90% de valores faltantes**.
  - `TransactionAmt` com cauda longa (distribuio lognormal).
  - `TransactionDT` usado como proxy temporal.

### Passos aplicados
- Criao de um identificador de usurio (`uid`) combinando `card1`, `addr1` e `dist1`.
- Seleo de um subconjunto enxuto de variveis para viabilizar a sequncia temporal.
- **Tratamento de missing values**:
  - Numricas  preenchimento pela mediana.
  - Categricas  `"missing"` + encoding inteiro.
- **Normalizao** das variveis numricas via `StandardScaler`.
- Ordenao por `uid` e `TransactionDT`.
- Construo de **sequncias temporais** de tamanho 10, rotulando cada janela pelo valor da ltima transao (`isFraud`).

---

## 2. Definio da Arquitetura do Modelo

- **Rede sequencial LSTM**:
  - `Masking` layer.
  - `LSTM(64, return_sequences=True)` + `Dropout(0.3)`.
  - `LSTM(32)` + `Dropout(0.3)`.
  - `Dense(64, relu)` + `Dropout(0.3)`.
  - `Dense(1, sigmoid)`.
- **Parmetros treinveis**: ~35k.
- **Funo de perda**: `binary_crossentropy`.
- **Otimizao**: Adam (lr=1e-3).
- **Mtrica principal**: AUC-ROC.

---

## 3. Treinamento e Validao

- Split estratificado 80/20.
- **Class weights** para lidar com desbalanceamento.
- **Callbacks**:
  - EarlyStopping (monitorando `val_AUC`).
  - ReduceLROnPlateau.

### Observaes durante o treinamento
- AUC de treino subiu continuamente (at ~0.86).
- AUC de validao estabilizou em torno de ~0.80 e comeou a cair  **sinal claro de overfitting leve**.
- As curvas de `loss` mostram melhora consistente, mas com um gap crescente entre treino e validao.

---

## 4. Resultados no Conjunto de Teste

- **Mtricas (threshold 0.5)**:
  - AUC-ROC: ~0.83
  - Preciso (fraude): ~0.12
  - Recall (fraude): ~0.65
  - F1 (fraude): ~0.20
- **Matriz de confuso**:
  - Muitos **falsos positivos** (347).
  - Poucos **falsos negativos** (26).
- **Curva ROC**: acima da aleatria, confirmando boa separabilidade.
- Problema identificado: **threshold padro de 0.5 no otimiza F1/Recall**, especialmente num dataset altamente desbalanceado.

---

## 5. Anlise Crtica

### Overfitting
- Diferena clara entre `train_AUC` e `val_AUC`.
- Modelo comea a memorizar padres especficos de treino aps poucas pocas.

### Estratgias aplicadas
- Dropout em vrias camadas.
- EarlyStopping e ReduceLROnPlateau.
- Class weights para balancear.
- Normalizao e encoding simples para reduzir rudo.

### Possveis melhorias
- **Threshold tuning** via curva PrecisionRecall para maximizar F1.
- **Focal Loss** para dar mais peso a exemplos difceis.
- Split mais robusto (por `uid` ou temporal) para evitar vazamento.
- Engenharia de features adicionais (`tempo desde ltima transao`, agregaes por usurio).
- Embeddings para categricas de alta cardinalidade.
- Arquiteturas alternativas (GRU, Attention, CNN-1D).

---
